```python
#Example of an object
class Person :

    def __init__(self,name,surname,emailid,dob):
        self.name1 = name
        self.surname = surname
        self.emailid = emailid
        self.dob = dob

nav_var = Person("Naveen","Kumar","naveen@gmail.com",4031988)

print(nav_var.name1)
print(nav_var.surname)
print(nav_var.emailid)
print(nav_var.dob)
```

```python
# Main Menu for Library Management System(LBMS) Version 1.0
import random
class Book:
    '''Book class for library management system'''
    BookCount = 0
    def __init__(self):
        self.title = input("Enter title of the Book")
        self.author = input("Enter author name")
        self.isbn = random.randint(0, 1000)
        self.pub_year = int(input("Enter the year of publication"))
        self.av_copies = int(input("Enter the number of available copies"))
        Book.BookCount = Book.BookCount + 1
    def display(self):
        print("Book Title\n",self.title,"\nAuthor",self.author,"\nISBN"+ str(self.isbn),"\nYear of Publication",self.pub_year,"\nCopies Available", self.av_copies)
    def add_to_library(self):
        self.total = 0
        self.total = self.total + self.av_copies
        print("Available copies",self.total)
    def ret_to_library(self):
        self.total = 0
        self.total =  self.av_copies - 1
        Book.BookCount = self.av_copies - 1
        print("After returning Books",self.total,"Book Count is",Book.BookCount)

def start():
    '''Menu Based Function for Library Management System(LBMS)'''
    while(True):
        print("Library Management Systems ")
        print("1. To Add a book to Library")
        print("2. To Return a book to Library")
        print("3. Enter to Display Status")
        print("4. Enter to Exit or Quit")
        try:
            choice = int(input("Enter a valid choice from 1-4"))
            print()
            if (choice == 1):
                bk = Book()
                bk.add_to_library()
            elif (choice == 2):
                bk.ret_to_library()
            elif (choice == 3):
                bk.display()
            elif (choice == 4):
                print("Thanks for using LBMS")
                break
            else:
                print("Invalid Choice!")
        except ValueError:
            print("Please enter valid choice as per Display")

start()
```
