```python
class Parent:
    def __init__(self, txt):
        self.message = txt

    def printmessage(self):
        print(self.message)

class Child(Parent):
      def __init__(self, txt):
        super().__init__(txt)

x = Child("Hello, and welcome!")

x.printmessage()
```

```python
class Vehicle:
    def __init__(self,make,model,year):
        self.make = make
        self.model = model
        self.year = year
    def display(self):
        print("-------------------------------")
        print("Make of Vehicle :",self.make)
        print("Model of Vehicle :",self.model)
        print("Year of Manufacture :",self.year)
        print("Type of Fuel :",self.fuel_type)
        print("--------------------------------")
        
        
class Car(Vehicle):
    def __init__(self,make,model,year,fuel_type):
        super().__init__(make,model,year)
        self.fuel_type=fuel_type
        
def start():
    while(True):
        make = input("Enter make of the Vehicle:")
        model = input("Enter model type :")
        year = int(input("Enter the year manufactured"))
        fuel_type = input("Enter the fuel type:")
        a=Car(make,model,year,fuel_type)
        a.display()
        break

start()
```

```python
class Employee:
    def __init__(self,name,salary):
        self.name = name
        self.salary = salary
    def display_manager(self):
        print("------------------------------------")
        print("Name of the Employee:",self.name)
        print("Salary of Employee:",self.salary)
        print("Department of Employee:",self.department)
        print("------------------------------------")
    def display_developer(self):
        print("------------------------------------")
        print("Name of the Employee:",self.name)
        print("Salary of Employee:",self.salary)
        print("Programming Language:",self.programming_language)
        print("------------------------------------")
        
class Manager(Employee):
    '''Derived class from Employee Base Class'''
    def __init__(self,name,salary,department):
        super().__init__(name,salary)
        self.department=department
       
        

class Developer(Employee):
    '''Derived class from Employee Base Class'''
    def __init__(self,name,salary,programming_language):
        super().__init__(name,salary)
        self.programming_language=programming_language
       
       
        
        
def start():
    while(True):
        try:
            name = input("Enter name of the employee:")
            salary = input("Enter salary of the employee :")
            choice = input("Is the employee Manager or Developer.Press M - Manager and D - Developer :")
            if choice.lower() in ["M","m"]:
                dept = input("Enter the department:")
                a=Manager(name,salary,dept)
                a.display_manager()
            elif choice.lower() in ["D","d"]:
                pl = input("Enter the programming language of developer:")
                b=Developer(name,salary,pl)
                b.display_developer()
            else :
                print("Enter a valid key!")
                break
        except ValueError:
            print("Invalid options")
                            
start()
```

```python
class Shape:
    def __init__(self,colour,border_width):
        self.colour=colour
        self.border_width = border_width
    
        
class Rectangle(Shape):
    '''Derived class from Shape Base Class'''
    def __init__(self,colour,border_width,length):
        super().__init__(colour,border_width)
        self.colour = colour
        self.length = length
        self.border_width = border_width
        print("Colour of the Rectangle is:",colour)
    def area(self,length,width):
        self.length = length
        self.border_width = width
        print("Area of the rectangle is",length*width)
        
class Circle(Shape):
    '''Derived class from Shape Base Class'''
    def __init__(self,colour,radius):
        super().__init__(colour,radius)
        self.colour = colour
        self.radius = radius
        print("Colour of the Circle is:",self.colour)
        
    def area(self,radius):
        self.radius = radius
        print("Area of the Circle is",radius*radius)
        
        
        
def start():
    while(True):
        try:
            choice = input("Press R-Rectangle and C-Circle:")
            if choice.lower() in ["R","r"]:
                length = int(input("Enter the length of the Rectangle:"))
                width = int(input("Enter the width of the Rectangle:"))
                colour = input("Colour of the Rectangle:")
                a=Rectangle(colour,length,width)
                a.area(length,width)
                print("Type of Shape is Rectangle")
            elif choice.lower() in ["C","c"]:
                radius = int(input("Enter the radius of the circle:"))
                colour = input("Enter the colour of the Circle:")
                b=Circle(colour,radius)
                b.area(radius)
                print("Type of Shape is Circle")
            else :
                print("Enter a valid key!")
                break
        except ValueError:
            print("Invalid options")
                            
start()          
```

```python
class Device:
    def __init__(self,brand,model):
        self.brand = brand
        self.model = model
    def display_phone(self):
        print("------------------------------------")
        print("Brand of the Phone:",self.brand)
        print("Model of the Phone:",self.model)
        print("Screen size of the Phone:",self.screen_size,"Inches")
        print("------------------------------------")
    def display_tablet(self):
        print("------------------------------------")
        print("Brand of the Tablet:",self.brand)
        print("Model of the Tablet:",self.model)
        print("Battery size of the Tablet is:",self.battery_size,"mAH")
        
        print("------------------------------------")

class Phone(Device):
    def __init__(self,brand,model,screensize):
        super().__init__(brand,model)
        self.screen_size = screensize
        
class Tablet(Device):
    def __init__(self,brand,model,battery):
        super().__init__(brand,model)
        self.battery_size = battery

def start():
    while(True):
        try:
            choice = input("Enter the device details P for Phone or T for Tablet:")
            if choice.lower() in ["P","p"]:
                brand = input("Enter the Brand of the Phone:")
                model = input("Enter the Model of the Phone:")
                size = float(input("Enter the screen size of the Phone:"))
                a=Phone(brand,model,size)
                a.display_phone()            
            elif choice.lower() in ["T","t"]:
                brand = input("Enter the Brand of the Tablet:")
                model = input("Enter the Model of the Tablet:")
                battery = int(input("Enter the Battery Size of the Tablet:"))
                b=Tablet(brand,model,battery)
                b.display_tablet()
            else :
                print("Enter a valid key!")
                break
        except ValueError:
            print("Invalid options")
                            
start()
```
